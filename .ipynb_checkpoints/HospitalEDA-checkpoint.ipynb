Breaking Down the "Measure Name" Codes:

HAI = Hospital-Acquired Infection

The Number (1, 2, etc.) = Type of Infection:

    HAI_1: Central Line-Associated Bloodstream Infections (CLABSI)

    HAI_2: Catheter-Associated Urinary Tract Infections (CAUTI)

    HAI_3: Surgical Site Infections (SSI)

    HAI_4: Methicillin-resistant Staphylococcus aureus (MRSA)

    HAI_5: Clostridioides difficile (C. diff)

    HAI_6: Ventilator-Associated Pneumonia (VAP)

The Suffix (e.g., CI_LOWER, NUMERATOR) = What the measure represents:

    CI_LOWER: (confidence interval) Lower bound of the expected/predicted infection rate range

    CI_UPPER: Upper bound of the expected/predicted infection rate range.

    NUMERATOR: Actual count of infections

    DOPC_DAYS: Days of Patient Care (denominator for infection rate calculation), Total patient-days/central-line days

    ELIGCASES: Eligible cases (patients meeting criteria for the measure) E.g 500 eligible patients (with central lines, long thin tube near the heart), 5 actual CLABSI infections

<-------------------------------------------  
<span style="color:#FF5733">i did some data wrangling, separation of string values just to get all unique types of HAIs in the dataset i'm working with.</span>

```python
# finding ALL HAIs covered in this dataset

print("\nUnique HAI measure types:")
hai_types = [name.split(':')[0].strip() for name in df['Measure Name'].unique() if ':' in name]
print(np.unique(hai_types))

# Alternative approach that might be more reliable
print("\nUnique HAI types (alternative method):")
# Extract HAI type from the measure name more explicitly
unique_hai_types = []
for measure_name in df['Measure Name'].unique():
    if ':' in measure_name:
        hai_type = measure_name.split(':')[0].strip()
        if hai_type not in unique_hai_types:
            unique_hai_types.append(hai_type)
print(unique_hai_types)

# # You can also check the Measure ID column for patterns
# print("\nUnique Measure ID prefixes:")
# measure_id_prefixes = [id.split('_')[0] + '_' + id.split('_')[1] for id in df['Measure ID'].unique()]
# print(np.unique(measure_id_prefixes))
```

Data Transformation to standardize datetime format to get the duration

```python
# Convert dates to datetime
df['Measure Start Date'] = pd.to_datetime(df['Measure Start Date'], format='%m/%d/%Y')
df['Measure End Date'] = pd.to_datetime(df['Measure End Date'], format='%m/%d/%Y')
```

```python
# Extract reporting period duration in days
df['Reporting Period Days'] = (df['Measure End Date'] - df['Measure Start Date']).dt.days
print(df['Reporting Period Days'])
```

```python
# Extract HAI type from Measure Name, feature extraction. Creating a new column from old data.
df['HAI Type'] = df['Measure Name'].apply(lambda x: x.split(':')[0] if ':' in x else x)
print(df['HAI Type'])
```

```python
# Extract measure parameter from Measure Name
df['Measure Parameter'] = df['Measure Name'].apply(lambda x: x.split(':', 1)[1].strip() if ':' in x else '')
print(df['Measure Parameter'])
```

## 1. What is the total number of observed HAI cases for each infection type?


<-------------  
<span style="color:#FF5733">important because it directly measures the weightage of each HAI in the entire country and sees if the country's HAI rates are above or below the national benchmark

```python
# Set plotting style
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = [10, 6]
```

* Number of Procedures	               = How many surgeries of a specific type were done. (E.g 500 colon surgeries performed, always a whole number)
* Observed Cases	                        = Actual number of HAI cases
* Predicted Cases	                        = Statistically expected number of infections based on models or national averages. 

* Lower/Upper Confidence Limit (LCL/UCL)	
The range of uncertainty around the predicted cases. Helps you judge whether your hospital is performing worse than expected.	
Based on model: Predicted infections = 10, LCL = 7, UCL = 13. So if your hospital had 15 observed cases, that’s higher than expected → might require investigation.

```python
df
```

```python
# Step 2: returns a df that only has rows /observed cases' under 'Measure Name' column
observed_cases = df[df['Measure Parameter'] == 'Observed Cases'].copy()
```

```python
observed_cases.info()
```

```python
observed_cases['Measure Name']
```

```python
# removing 'Not available' values under 'Score' column. replacing these 'not available' values with 0.0
observed_cases['Score'] = pd.to_numeric(observed_cases['Score'], errors='coerce').fillna(0)
print(observed_cases['Score'])
```

```python
# Step 3: Calculate the total observed cases by HAI type. score means 2 patients / 1000 patient days (total number patients for a set number of days ____ need calcualte measure start and end date)
# we group by HAI type and get the count of each HAI cases
total_by_hai_type = observed_cases.groupby('HAI Type')['Score'].sum().sort_values(ascending=False)
print(total_by_hai_type)
```

<----------------------------  
<span style="color:#FF5733">
the code above was the result of some data cleaning, I eventually managed to get the number of HAI cases for each HAI type respectively  
unfortunately, even though this dataset has other types:  
'CAUTI', 'CLABSI', 'SSI', 'Surgical Site Infection from abdominal hysterectomy (SSI), 'Surgical Site Infection from colon surgery (SSI)'  
only CAUTI HAI, CLABSI HAIs are reported to have Observed Cases (confirmed cases), while SSI, abdonminal surgeyr, colon surgery only have data for Upper Lower Limits, number of surgery procefures and predicted cases.

```python
plt.figure(figsize=(12, 7))
ax = total_by_hai_type.plot(kind='bar', color='tomato')
plt.title('Total Number of Observed HAI Cases by Infection Type', fontsize=14)
plt.xlabel('Infection Type')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of each bar
for i, v in enumerate(total_by_hai_type):
    plt.text(i, v + 5, f'{int(v)}', ha='center')

plt.tight_layout()
```

```python
earliest_start = df['Measure Start Date'].min()
latest_end = df['Measure End Date'].max()
print(f"number of CAUTI and CLABSI from {earliest_start} to {latest_end}")
```

```python
# getting predicted values for CAUTI and CLABSI infections
predicted_cases = df[df['Measure Parameter'] == 'Predicted Cases'].copy()
predicted_cases['Score'] = pd.to_numeric(predicted_cases['Score'], errors='coerce').fillna(0)
total_predicted_by_type = predicted_cases.groupby('HAI Type')['Score'].sum().reindex(total_by_hai_type.index)
print(total_predicted_by_type)
```

```python
# just creating a simple df to compare
comparison_df = pd.DataFrame({
    'Observed Cases': total_by_hai_type,
    'Predicted Cases': total_predicted_by_type
})
comparison_df
```

```python
plt.figure(figsize=(12, 7))
comparison_df[['Observed Cases', 'Predicted Cases']].plot(kind='bar')
plt.title('Comparison of Observed vs Predicted HAI Cases by Infection Type', fontsize=14)
plt.xlabel('Infection Type')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
```

```python
# calculating the Standardized Infectionn Ratio (SSI)
# Calculate the ratio of observed to predicted (O/E ratio)
comparison_df['SIR Ratio'] = (comparison_df['Observed Cases'] / comparison_df['Predicted Cases']).round(2)
print("\nSIR Ratio by HAI Type:")
print(comparison_df['SIR Ratio'])
```

<-------------------------------   
<span style="color:#FF5733">So we can finally answer the question "What is the total number of observed HAI cases for each infection type?" between '2014-07-01' &' 2015-06-30', approximately 1 year worth of data, there are 11,189 observed cases of Catheter-Associated Unrinary Tract Infection (CAUTI) & 9,319 observed cases of Central Line Associated Bloodstream Infection (CLABSI).

<span style="color:#FF5733">And since both HAI's ratio is < 1, it shows that in all hospitals collectively, the national average of CAUTI and CLABSI HAI rate is lower than expected. the expected being the national average HAI rate taking into account the countries' hospital patient vollume, number of days devices (catheter or ventilator) is used on patients and so on.

<span style="color:#FF5733">we can also see that the country is predicting ~50% more than the country's actual observed infections.

## 2. Is there a relationship between the number of procedures performed and the standardized infection ratio (SIR)?


<--------------  
<span style="color:#FF5733">  
this question is important when analyzing "volume-risk relationships". the numbber of patient day care/ procedures represents patient's exposure to infection risks.

<span style="color:#FF5733">Otherwise, we risk underestimating the infection risk in high-volume wards or hospitals.

<span style="color:#FF5733">If SIR increases with procedure count, we need to consider infection control measures need to scale proportionally.

```python
num_procedures = df[df['Measure Parameter'] == 'Number of Procedures'].copy()
num_procedures
```

```python
# just to check if there might be other HAIs. but no only CAUTI and CLABSI has Observed Cases.
# just to check if there's 'Not available' values'
print(num_procedures['HAI Type'].unique())
print(num_procedures['Score'].unique())
```

```python
# get the number of procedures (TOTAL) for CAUTI and CLABSI each
num_procedures['Score'] = pd.to_numeric(num_procedures['Score'], errors='coerce').fillna(0)
```

```python
total_procedures = num_procedures.groupby("HAI Type")["Score"].sum()
total_procedures
```

```python
print(type(comparison_df['SIR Ratio']))
```

```python
# combine series into 1 dataframe
combined_df = pd.concat([total_procedures, comparison_df['SIR Ratio']], axis=1)
combined_df
```

score == number of procedures == DAYS of patient care (number of days in hospital being cared for) for CAUTI and CLABSI

```python
fig, ax1 = plt.subplots(figsize=(8, 5))

color1 = 'tab:blue'
color2 = 'tab:orange'

# Bar for number of procedures
ax1.set_xlabel("HAI Type")
ax1.set_ylabel("Number of Procedures", color=color1)
bars = ax1.bar(combined_df.index, combined_df["Score"], color=color1, alpha=0.6)
ax1.tick_params(axis='y', labelcolor=color1)

# ✅ Add values on top of the bars
for bar in bars:
    height = bar.get_height()
    ax1.annotate(f'{height:,.0f}',  # formats the number with commas
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5),  # vertical offset
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=9, color='black')

# Second y-axis for SIR
ax2 = ax1.twinx()
ax2.set_ylabel("SIR Ratio", color=color2)
ax2.plot(combined_df.index, combined_df["SIR Ratio"], color=color2, marker='o', linewidth=2)
ax2.tick_params(axis='y', labelcolor=color2)

plt.title("Procedures vs SIR for CAUTI and CLABSI")
plt.tight_layout()
plt.show()
```

<------------------------------------------  
To answer the question "Is there a relationship between the no. of days of patient care(number of procedures performed) and the standardized infection ratio (SIR)?". 

there appears to be a slight positive relationship between the number of procedures performed and the SIR ratio — specifically for CAUTI. CAUTI, which had a higher number of procedures (over 10 million), also had a slightly higher SIR (0.57) compared to CLABSI (0.56), which had fewer procedures (about 9.7 million)

from this small dataset and 2 data points, the results cannot be considered statistically significant.
